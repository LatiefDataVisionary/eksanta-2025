## **Mengimpor Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("zara.csv", sep=';')
df.head()
```

Promotion: Menunjukkan apakah produk dipromosikan atau tidak

Product Category Kategorinya adalah pakaian

Brand Merek

Deskription: Deskripsi produk.

Terms: Tipe dari pakaian

Section: Menunjukkan Bagian misalnya, pakaian wanita, pakaian pria



## **2. Data Cleaning dan Data Preprocessing**

```python
df.isnull().sum()
```

```python
df = df.dropna(subset= ['name','description']) #Menghilangkan baris yang terdapat value kosongnya(null)
df
```

```python
# saya berfikir bahwa data seperti "url","sku","description", dan "scraped_at" tidak berkorelasi kemanapun
# jadi saya hilangkan
df = df.drop(['url','sku','description','scraped_at'],axis=1)
df.head()
```

```python
df['total price'] = df['price'] * df['Sales Volume']
df.head()
```

## **3. Exploratory Data Analysis (EDA)**

```python
# Menghitung Matrik Korelasi
matriks_korelasi = df.select_dtypes(include=np.number).corr()

# Membuat visualisasi matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(matriks_korelasi, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriks Korelasi')
plt.show()
```

### **a. Univariate Analysis:**


Distribusi Price dan Sales Volume

```python
# Distribusi kolom 'price'
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribusi kolom "price"')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Distribusi kolom 'Sales Volume'
plt.figure(figsize=(8, 6))
sns.histplot(df['Sales Volume'], kde=True)
plt.title('Distribusi kolom "Sales Volume"')
plt.xlabel('Volume Penjualan')
plt.ylabel('Frekuensi')
plt.show()
```

Eksplorasi Kolom Kategorikal

```python
df['terms'].value_counts()
```

```python
# Bar chart utk kolom 'terms' (tipe produk)
plt.figure(figsize=(10, 6))
sns.countplot(x='terms', data=df)
plt.title('Frekuensi Tipe Produk')
plt.xlabel('Tipe Produk')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

```python
df['section'].value_counts()
```

```python
# Bar chart utk kolom 'section' (gender pakaian)
plt.figure(figsize=(8, 6))
sns.countplot(x='section', data=df)
plt.title('Distribution Across Store Sections')
plt.xlabel('Store Section')
plt.ylabel('Count')
plt.show()
```

### **b. Bivariate Analysis**

Harga vs Volume Penjualan

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='price', y='Sales Volume', data=df)
plt.title('Harga vs Volume Penjualan')
plt.xlabel('Harga')
plt.ylabel('Volume Penjualan')
plt.show()
```

Volume Penjualan Berdasarkan Tipe Produk

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='terms', y='Sales Volume', data=df)
plt.title('Volume Penjualan Berdasarkan Tipe Produk')
plt.xlabel('Tipe Produk')
plt.ylabel('Volume Penjualan')
plt.xticks(rotation=45)
plt.show()
```

Volume Pendapatan Rata-Rata Berdasarkan Jenis Produk

---



```python
df.groupby('terms')
```

```python
cdf= df.groupby('terms')['total price'].mean()
cdf
```

```python
# Volume Pendapatan Rata-Rata Berdasarkan Jenis Produk
plt.figure(figsize=(8, 6))
cdf.plot(kind='line', color='red')
plt.title('Volume Pendapatan Rata-Rata Berdasarkan Jenis Produk')
plt.xlabel('Tipe Produk')
plt.ylabel('Rata-rata Pengeluaran')
plt.show()
```

Perhitungan pendapatan untuk setiap produk yang dipromosikan atau tidak dipromosikan

```python
ax= df.groupby('Promotion')['total price'].sum()
ax
```

```python
# Perhitungan pendapatan untuk setiap produk yang dipromosikan atau tidak dipromosikan
plt.figure(figsize=(8, 5))
colors= ['#FF0000','#0090FF']
ax.plot(kind ='pie',autopct='%.1f%%', colors=colors)
plt.title('Pendapatan produk berdasarkan promosi')
plt.ylabel(' ')
plt.show()
```

```python
# memilih volume penjualan terhadap posisi produk dan menjumlahkan nilai penjualan yang dihasilkan
adf = df.groupby('Product Position')['Sales Volume'].sum()
adf
```

```python
# Membuat bar chart
plt.figure(figsize=(8, 5))
adf.plot(kind='bar', color=['skyblue', 'lightgreen', 'LightSalmon'])

# Kustomisasi plot
plt.title('Volume Penjualan Berdasarkan Posisi Produk')
plt.xlabel('Product Position')
plt.ylabel('Volume Penjualan')
plt.xticks(rotation = 360)
plt.show()
```

```python
# Mengelompokkan data berdasarkan 'name' dan menghitung rata-rata 'total price'
top= df.groupby('name')['total price'].mean()
tp=top.head(5)
tp
```

```python
# Membuat line chart
plt.figure(figsize=(10, 8))
tp.plot(marker='o', linestyle='-', color='blue')

# Kustomisasi plot
plt.title('Rata-rata Harga Berdasarkan Nama Produk (Top 10)')
plt.xlabel('Nama Produk')
plt.ylabel('Harga Rata-rata')
plt.xticks(rotation=45)  # Memutar label sumbu X jika diperlukan
plt.grid(True)
plt.show()
```

```python

# Mendapatkan produk termahal
produk_termahal = df.nlargest(5, 'price')

# Menentukan irisan termahal
explode = [0.1 if i == produk_termahal['price'].idxmax() else 0 for i in produk_termahal.index]

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(produk_termahal['price'], labels=produk_termahal['name'], autopct='%1.2f%%', colors=sns.color_palette('deep'), explode=explode)

# Kustomisasi plot
plt.title('Produk Termahal')
plt.axis('equal')  # Untuk memastikan pie chart dalam bentuk lingkaran
plt.show()
```

```python
# Mendapatkan produk termurah
produk_termurah = df.nsmallest(5, 'price')

# Menentukan irisan termurah
explode = [0.1 if i == produk_termurah['price'].idxmax() else 0 for i in produk_termurah.index]

# Membuat pie chart
plt.figure(figsize=(7, 7))
plt.pie(produk_termurah['price'], labels=produk_termurah['name'], autopct='%1.2f%%', colors=sns.color_palette('pastel'), explode=explode)

# Kustomisasi plot
plt.title('Produk Termahal')
plt.axis('equal')  # Untuk memastikan pie chart dalam bentuk lingkaran
plt.show()
```

```python
# menemukan jumlah penjualan terhadap bagian pengguna
X= df.groupby('section')['Sales Volume'].sum()
X
```

```python
plt.figure(figsize=(10, 8))
X.plot(kind='pie', autopct='%.1f%%', colors=['green', 'yellow'])
plt.title('Distribusi Volume Penjualan berdasarkan Bagian Toko', fontsize= 16)
plt.ylabel(' ')
plt.show()
```
